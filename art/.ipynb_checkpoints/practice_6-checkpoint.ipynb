```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/test.jpg', 0)
```

```python
rows, cols = img.shape
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)
```

```python
M = np.float32([
        [1, 0, 0],
        [0.5, 1, 0],
        [0, 0, 1]
])
sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))
```

```python
plt.subplot(1,2,2)
plt.title('Sheared image - Y axis')
plt.imshow(sheared_img)
```

```python
cv.imwrite('images/sheared_out_y.jpg', sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
