```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/test.jpg', 0)
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)
```

```python
cropped_img = img[250:400, 750:1500]
```

```python
plt.subplot(1,2,2)
plt.title('Cropped image')
plt.imshow(cropped_img)
```

```python
cv.imwrite('images/cropped_out.jpg', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
