```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/test.jpg', 0)
```

```python
rows, cols = img.shape
```

```python

rotated_img = cv.warpAffine(
    img, 
    cv.getRotationMatrix2D(
        (cols/2, rows/2), 
        30,
        0.6
    ),
(cols, rows))
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)
```

```python
plt.subplot(1,2,2)
plt.title('Rotated image')
plt.imshow(rotated_img)
```

```python
cv.imwrite('images/rotated_out.jpg', rotated_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
