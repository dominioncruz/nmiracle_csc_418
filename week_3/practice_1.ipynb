```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/test.jpg', 0)
```

```python
rows, cols = img.shape
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)
```

```python
M = np.float32([
        [1, 0, 100],
        [0, 1, 50]
])
transl = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.subplot(1,2,2)
plt.title('Translated image')
plt.imshow(transl)
```
