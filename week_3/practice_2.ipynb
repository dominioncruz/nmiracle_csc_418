```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/test.jpg', 0)
```

```python
rows, cols = img.shape
```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)
```

```python
M = np.float32([
        [1, 0, 0],
        [0, -1, rows],
        [0, 0, 1]
])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))
```

```python
plt.subplot(1,2,2)
plt.title('Reflected image')
plt.imshow(reflected_img)
```

```python
cv.imwrite('images/reflected_out.jpg', reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
