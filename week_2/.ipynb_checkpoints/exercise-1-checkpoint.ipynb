```python
def get_member_info(member_email):
    for member, info in members.items():
        if member.split()[1].lower() == member_email:
            return member, info
    return None


def compare_password(provided_password, actual_password):
    count = sum([i for i in range(1, len(actual_password) + 1)])
    return provided_password == str(count)
```

```python
email = str(input("Please enter your email: ")).lower().strip()
password = str(input("Please enter your password: "))

details_correct = False

result = get_member_info(email)
if result:
    member, info = result
    if compare_password(password, member.split()[2]):
         details_correct = True

if details_correct:
    image = cv2.imread('members/{}.jpg'.format(email))
    rgb_version = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.subplot(1,2,1)
    plt.title('{} - {}'.format(member, info['position']))
    plt.imshow(rgb_version)
    plt.axis('off')
    plt.show()

    print(info['about'])
else:
    print('invalid email or password')
```
