```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/test.jpg')

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image)

kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

sharpened_image = cv2.filter2D(image, -1, kernel)

cv2.imwrite('images/sharpened_test.jpg', sharpened_image)

plt.subplot(1,2,2)
plt.title('Sharpening')
plt.imshow(sharpened_image)
plt.show()
```
