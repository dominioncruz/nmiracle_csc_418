```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/test.jpg')

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image)

filtered_image = cv2.medianBlur(image, 5)

cv2.imwrite('images/blur_test.jpg', filtered_image)

plt.subplot(1,2,2)
plt.title('Median blur')
plt.imshow(filtered_image)
plt.show()
```
