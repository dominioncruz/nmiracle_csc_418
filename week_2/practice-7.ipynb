```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/test.jpg')

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image)

inverse_image = 255 - image

cv2.imwrite('images/inverse_test.jpg', inverse_image)

plt.subplot(1,2,2)
plt.title('Inverse image')
plt.imshow(inverse_image)
plt.show()
```
