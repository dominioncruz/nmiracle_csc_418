```python
import cv2

import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/test.jpg')

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(image)

scaled_image = cv2.resize(image, None, fx=2, fy=2)

cv2.imwrite('images/scaled_test.jpg', scaled_image)

plt.subplot(1,2,2)
plt.title('Scaled')
plt.imshow(scaled_image)
plt.show()
```
